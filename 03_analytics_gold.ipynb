### Camada Ouro (Gold)

É nesta etapa que os dados deixam de ser **"tabelas de sistema"** e viram **"indicadores de negócio"**

Neste notebook, nosso objetivo vai ser responder a uma pergunta clássica de negócio: **"Quais são os estados brasileiros que mais geram faturamento real para a empresa?"**

```python
%sql
CREATE OR REPLACE TABLE gold_olist_revenue_by_state
USING DELTA
AS
SELECT 
    c.state AS uf,
    COUNT(DISTINCT o.order_id) AS qtd_pedidos,
    SUM(i.price) AS faturamento_produtos,
    SUM(i.freight_value) AS faturamento_frete,
    (SUM(i.price) + SUM(i.freight_value)) AS faturamento_total
FROM silver_olist_orders o
INNER JOIN silver_olist_customers c 
    ON o.customer_id = c.customer_id
INNER JOIN silver_olist_order_items i 
    ON o.order_id = i.order_id
WHERE o.order_status = 'delivered'
GROUP BY c.state
ORDER BY faturamento_total DESC;
```
